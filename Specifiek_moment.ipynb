## gem_freq_u_day | gem_freq_u_ev | gem_freq_u_night

```python
gem_freq_d = df.groupby(['id', 'dag', 'uur' ])['fancyname'].count().groupby(['id', 'uur']).mean().to_frame().reset_index()
```

```python
gem_freq_d
```

Hoeveel gemiddeld geen smartphonegebruik 's nachts

```python
rust = pd.pivot_table(gem_freq_d, values='fancyname', index=['id'],
                    columns=['uur'])
```

```python
rust.fillna(0)
```

```python
rust = rust.reset_index()
```

```python
rust = rust.fillna(0)
```

```python
rust
```

```python
rust_long =pd.melt(rust,id_vars=['id'],var_name='uren', value_name='mean appE')
```

```python
rust_long = rust_long.sort_values(by=['id', 'uren'])
```

```python
rust_long.head()
```

```python
def avondgebruik(row):
    uur = row["uren"]
    
    if (uur >= 6) & (uur < 21):
        return "overdag gebruik"
    elif (uur >= 21) & (uur <= 23):
        return "avondgebruik"
    elif (uur >= 0) & (uur < 6):
        return "nachtgebruik"
    else:
        return "niet gevonden"
```

```python
rust_long['gebruik'] = rust_long.apply(avondgebruik, axis=1)
```

```python
rust_long.head()
```

```python
rust_long_sum = rust_long.groupby(['id','gebruik' ])['mean appE'].sum().reset_index()
```

```python
rust_long_sum.head()
```

```python
rust_long_sum.columns = ['id', 'gebruik', 'sum appE']
```

```python
def meanU(row):
    gebruik = row["gebruik"]
    value = row["sum appE"]
    
    if (gebruik == 'avondgebruik'):
        return value/3
    elif (gebruik == 'nachtgebruik'):
        return value/6
    elif (gebruik == 'overdag gebruik'):
        return value/15
    else:
        return "niet gevonden"
```

```python
rust_long_sum['mean appE'] = rust_long_sum.apply(meanU, axis=1)
```

```python
rust_long_sum.head()
```

```python
gem_freq_u = pd.pivot_table(rust_long_sum, values='mean appE', index=['id'],
                    columns=['gebruik'])
```

```python
gem_freq_u
```

```python
gem_freq_u = gem_freq_u.reset_index()
```

```python
gem_freq_u.columns = ['id','gem_freq_u_day', 'gem_freq_u_ev', 'gem_freq_u_night']
```

```python
gem_freq_u.head()
```

```python
gem_freq_u.to_excel('gem_freq_u.xlsx')
```
