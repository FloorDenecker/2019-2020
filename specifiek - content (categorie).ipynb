## per_d_duur_...

###### = gemiddelde percentage van dagelijkse duur smartphonegebruik gelinkt met categorie

```python
min_id_dag_cat = df.groupby(['id', 'dag', 'category_v1'])['duur_min'].sum()
```

```python
min_id_dag_cat = min_id_dag_cat.to_frame().reset_index()
```

```python
min_id_dag_cat.head()
```

```python
min_id_dag = df.groupby(['id', 'dag'])['duur_min'].sum().to_frame().reset_index()
```

```python
min_id_dag.head()
```

```python
per_d_duur = pd.merge(min_id_dag_cat, min_id_dag, on = ['id', 'dag'])
```

```python
per_d_duur.head()
```

```python
per_d_duur.columns = ['id', 'dag', 'category_v1', 'min_cat_dag', 'min_dag']
```

```python
per_d_duur['per_d_duur'] = per_d_duur['min_cat_dag']/per_d_duur['min_dag']
```

```python
per_d_duur_cat = per_d_duur.groupby(['id', 'category_v1'])['per_d_duur'].mean().to_frame()
```

```python
per_d_duur_cat = per_d_duur_cat.reset_index()
```

```python
per_d_duur_cat.head()
```

```python
per_d_duur_categories = pd.pivot_table(per_d_duur, values='per_d_duur', index=['id'],
                    columns=['category_v1'])
```

```python
per_d_duur_categories_need = per_d_duur_categories[['social','communication', 'tools', 'productivity', 'video players & editors', 'music & audio', 'news & magazines', 'entertainment', 'game']]
```

```python
per_d_duur_categories_need.columns = ['per_d_duur_social','per_d_duur_com','per_d_duur_tools','per_d_duur_prod','per_d_duur_video','per_d_duur_music','per_d_duur_news','per_d_duur_entertain','per_d_duur_games']
```

```python
per_d_duur_categories_need.to_excel('per_d_duur_cat.xlsx')
```
